```python
import math as m
import numpy as np
import matplotlib.pyplot as plt

###################################
def f(t, y):      #es y'
    #return t*y                                                  #editar#
    #return 2*m.exp(-5*t)-0.9*y
    return -2*m.exp(-2*t)-3*m.exp(-3*t)+m.cos(t)+m.sin(t)
###################################
def y_real(t):
    return m.exp((t**2)*(0.5))                                   #editar#
###################################

def euler(a,b,t0,y0,h):
    n= int((b - a)/ h)
    t = [t0]
    y = [y0]
    fy = []
    error = []
    yreal = []
    error_rel = []

    for i in range(n):
        t.append(t[i] + h)
        y.append(round(y[i] + h* f(t[i], y[i]), 5))             #para tener mas decimales
        error.append(abs(y_real(t[i])- y[i]))

    for i in range(n+1):
        fy.append(f(t[i],y[i]))                                  
        yreal.append(y_real(t[i]))
        error_rel.append(abs(y[i] - yreal[i])/yreal[i])

    
        #if i==3 or i==4:
         #   print(f"El error en t={i} es:", error[i])

    print("t =", t)
    print("y =", y)
    print("y' =", fy)
    print("y_real =", yreal)
    print("error global =", error)
    print("error_rel =", error_rel)

    plt.xlim(0, 4)
    plt.ylim(0, 4)
    plt.scatter(t, y)
    plt.plot(t, y, c="green")

    x = np.linspace(a, b, 100)
    y_r = [y_real(i) for i in x] 

    plt.plot(x, y_r, c="red")

######################
euler(3, 4, 0, 0, 0.1) #a, b, t0, y0, h
#tk = a + k*h
######################
```
