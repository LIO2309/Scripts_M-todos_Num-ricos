```python
def biseccion(f,a,b,e):

    error = b - a
    k = 0

    c_aprox = []
    while error > e:
        c = (a + b)/2
        c_aprox.append(c)

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

        error = (b-a)/2

        k += 1

        print(c, error)
    print("numero de iteraciones", k)

################################################################################

def f(x):
    return (x**2)-m.sin(x)-0.5

biseccion(f, 0, 2, 10**-3) # var 2 y 3: lims de intervalo ; var 4: cota de error

################################################################################
```

```python
def newton(f,f_prima, x0, e):

    error = 1
    x =[x0]

    k = 0
    while error > e :

        xk = x[k] - ( f(x[k]) / f_prima(x[k]))
        x.append(xk)

        error = abs(f(x[k]) / f_prima(x[k]))

        k += 1

        print(xk, error)
    print("El numero de iteraciones es:", k)

####################################

def f(x):
    #return m.exp(x) - m.sin(x) #f
    return (x**3)+2.967*(x**2)-9.46693*x-11.557935
    
def f_prima(x):
    #return m.exp(x) - m.cos(x) #f'
    return 3*(x**2)+5.934*x-9.46693
    
newton(f, f_prima, 2, 0.00000000001) # var 3: valor inicial ; var 4: cota de error

####################################
#biseccion(f, -4, -3, 0)
```
